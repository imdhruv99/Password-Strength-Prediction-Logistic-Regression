**seprating independent and dependent feature**

```python
password_tuple = np.array(data)
```

```python
# suffling data so it will provide robustness to my model
import random

random.shuffle(password_tuple)
```

```python
# seprating independent and dependent using list comprehension
x = [labels[0] for labels in password_tuple] 
y = [labels[1] for labels in password_tuple] 
```

```python
x
```

```python
y
```

### Passing Data to TF-IDF

```python
# converting data into characters

def word_divide_char(inputs):
    character = []
    for i in inputs:
        character.append(i)
    return character
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# vectorizing data and achieving matrix
vectorizer = TfidfVectorizer(tokenizer = word_divide_char)
```

```python
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
# getting features of our data
vectorizer.get_feature_names()
```

```python
# object of vector
first_document_vector = X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
# creating dataframe
df = pd.DataFrame(first_document_vector.T.todense(), index = vectorizer.get_feature_names(), columns=['TF-IDF'])
```

```python
df.head()
```

```python
# sorting data on basis of TF-IDF
df.sort_values(by=['TF-IDF'], ascending=False)
```

- now we can see that each and every term as it's own frequency

### Feeding our processed data to our Logistic Regression model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# multi_class='multinomial' because we have more than two subcategory [0, 1, 2] 
model = LogisticRegression(random_state=0, multi_class='multinomial')
```

```python
model.fit(X_train, y_train)
```

#### Prediction on Data

```python
dt = np.array(['%@531wu_247#'])
pred = vectorizer.transform(dt)
model.predict(pred)
```

```python
y_pred = model.predict(X_test)
y_pred
```

### Performance of Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

**Accuracy of model is: 82%**
